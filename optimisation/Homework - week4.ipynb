```python
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

### Paraboloid function

```python
# implement our function
f = lambda x, y: 1/2*x**2 + 7/2*y**2
```

```python
x = np.arange(-1, 1, 0.01)
y = np.arange(-3, 3, 0.03)

X, Y = np.meshgrid(x, y)
Z = f(X,Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, cmap=cm.gist_heat_r)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

The function:
$$
f(x, y) = \frac{1}{2}x^2+\frac{7}{2}y^2
$$

The gradient:
$$
\nabla f = \left[ \frac{\partial f}{\partial x},\frac{\partial f}{\partial y} \right] = \left[x ,7y \right]
$$

It is easy to see that $f(x,y)\geq0$ and a global minimum of the function is $f(x^*,y^*)=0$ when $(x^*,y^*)=(0,0)$. Of course we can see from the plot and the second order condition as well that $f$ is a convex function.

```python
# implement the gradient of our function
f_grad_x = lambda x: x
f_grad_y = lambda y: 7*y
```

```python
# simple implementation of the gradient descent algorithm step
def gradient_descent(x0, y0, eta):
    x1 = x0 - eta*f_grad_x(x0)
    y1 = y0 - eta*f_grad_y(y0)
    return x1, y1
```

```python
# initial coordinates
x, y = (5, 5)

# learning rate
eta = 0.1

# number of steps
steps = 40

for i in range(steps+1):
    f_xy = f(x, y)
    print(f'Step{i}, eta={eta}, (x,y)=({x:.2f}, {y:.2f}), f(x,y)={f_xy:.2f}')
    x, y = gradient_descent(x, y, eta)
```

### Rosenbrock's function

```python
b = 100;
f = lambda x, y: (x-1)**2 + b*(y-x**2)**2;
```

```python
# Evaluate function
x = np.arange(-1, 2, 0.15)
y = np.arange(-1, 2, 0.15)

X, Y = np.meshgrid(x, y)
Z = f(X,Y)

# Initialize figure 
fig = plt.figure(figsize=(12, 7))
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, cmap=cm.gist_heat_r)
ax.set_zlim(0, 1000)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

The function:
$$
f(x, y) = 100(y-x^2)^2+(1-x)^2
$$

The gradient:
$$
\nabla f = \left[ \frac{\partial f}{\partial x},\frac{\partial f}{\partial y} \right] = \left[-400x(y-x^2)-2(1-x) ,200(y-x^2) \right]
$$

The global minimum of the function is $f(x^*,y^*)=0$ when $(x^*,y^*)=(1,1)$. We can see from the plot that $f$ is not a convex function.

```python
# implement the gradient of our function
f_grad_x = lambda x, y: -400*x*(y-x**2)-2*(1-x)
f_grad_y = lambda x, y: 200*(y-x**2)
```

```python
# simple implementation of the gradient descent algorithm step
def gradient_descent(x0, y0, eta):
    x1 = x0 - eta*f_grad_x(x0, y0)
    y1 = y0 - eta*f_grad_y(x0, y0)
    return x1, y1
```

```python
# initial coordinates
x, y = (6, 6)

# learning rate
eta = 5*10**(-5)

# number of steps
steps = 20

for i in range(steps+1):
    f_xy = f(x, y)
    print(f'Step{i}, eta={eta}, (x,y)=({x:.2f}, {y:.2f}), f(x,y)={f_xy:.2f}')
    x, y = gradient_descent(x, y, eta)
```
