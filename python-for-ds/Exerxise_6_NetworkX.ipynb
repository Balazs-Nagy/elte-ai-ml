### The `data/brain_small.graphml` file is a graph abstraction of a patient's brain. For the following problems, consult the NetworkX documentation if necessary

# 1.a

Load the data with NetworkX, access node and edge attributes (https://networkx.org/documentation/networkx-1.10/reference/generated/networkx.classes.function.get_node_attributes.html), and return the node name (dn_name attribute) for the node with the largest degree.

```python
# Solution
g = nx.read_graphml('data/brain.graphml')
node, degree = sorted(g.degree(), key=lambda x: x[1], reverse=True)[0]
print('The node name (dn_name attribute) for the node with the largest degree:')
print(nx.get_node_attributes(g, 'dn_name').get(node))
```

# 1.b 
Return the largest connected component of the network.

```python
# Solution
c = [c for c in sorted(nx.connected_components(g), key=len, reverse=True)]
print('Nodes of the largest connected component of G:')
print(c[0])
```

# 2.

For the node with name "lh.insula_9", find the shortest path to the other nodes with respect to the "number_of_fibers" edge attribute.
From these, find the target node which has the highest weight (by the number_of_fibers edge weight).
Print the weight of this path and the path.


Hint: Use the nx.shortest_path

```python
# Solution
d = nx.get_node_attributes(g, 'dn_name')
ivd = {v: k for k, v in d.items()}
source = ivd.get('lh.insula_9')
source
```

```python
shortest_path = nx.shortest_path(G=g, source=source, weight='number_of_fibers', method='dijkstra')
shortest_path
```

```python
shortest_path_length = nx.shortest_path_length(G=g, source=source, weight='number_of_fibers', method='dijkstra')
shortest_path_length
```

```python
node, weight_of_path = sorted(shortest_path_length.items(), key=lambda x: x[1], reverse=True)[0]

print('The path with the largest weight:')
print(shortest_path.get(node))
print('The weight of this path:')
print(weight_of_path)
```

```python
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D

def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

n = 100

# For each set of style and range settings, plot n random points in the box
# defined by x in [23, 32], y in [0, 100], z in [zlow, zhigh].
for c, m, zlow, zhigh in [('r', 'o', -50, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 32)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zlow, zhigh)
    ax.scatter(xs, ys, zs, c=c, marker=m, s=100)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
x = np.array(list(nx.get_node_attributes(g, 'dn_position_x').values()))
y = np.array(list(nx.get_node_attributes(g, 'dn_position_y').values()))
z = np.array(list(nx.get_node_attributes(g, 'dn_position_z').values()))

hemisphere = list(nx.get_node_attributes(g, 'dn_hemisphere').values())
mapping = { 'right': 'r', 'left': 'b'}
hemisphere = [h.replace(h, mapping.get(h)) for h in hemisphere]
```
