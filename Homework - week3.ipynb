```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# create 1000 equally spaced points between -1 and 5
x = np.linspace(-1, 5, 1000)

# choose lambda of the Lagrangian
l = 2

# calculate the points of the parabolas
y = lambda x: (x-2)*(x-4)
z = lambda x: x**2+1

# calculate the points of the Lagrangian
lagrangian = lambda x, l: z(x) + l*y(x)

# create plot
fig, ax = plt.subplots()

# draw coordinate lines
ax.axhline(y=0, color='black')
ax.axvline(x=0, color='black')

# draw curves
ax.plot(x, y(x), label='$y(x)=(x-2)(x-4)$')
ax.plot(x, z(x), label='$z(x)=x^2+1$')
ax.plot(x, lagrangian(x, l), label='Lagrangian: $L(x,\lambda)=z(x)+\lambda y(x)$ for $\lambda=$' + str(l))

# calculate and fill the feasible area
x_feasible = np.linspace(2, 4, 1000)
plt.fill_between(x_feasible, y(x_feasible), 0, step="pre", alpha=0.4)

# mark the optimal solution
plt.plot(2, z(2), 'ro', label='optimal value and solution $z(2)=5$')
plt.vlines(2, 0, z(2), color='r', linestyle="dashed")
plt.hlines(z(2), 0, 2, color='r', linestyle="dashed")

# misc. plot settings
fig.set_size_inches(18.5, 10.5)
plt.legend(loc='upper right')
plt.grid(True)
plt.title('Highlighted area shows feasible set')

# show the plot
plt.show()
```

Let $z(x)$ denote the objective function
$$z(x)=x^2+1$$

and $y(x)$ denote the constraint

$$y(x)=(x-2)(x-4)\leq0$$

Using the plots above we can easily observe that in case of the primal problem the **optimal value** of $x$ is $2$ and the **optimal solution** (minimum value of the objective function) is $z(2)=2^2+1=5$.

The Lagrangian $L(x,\lambda)$ can be written as follows:

$$L(x,\lambda)=z(x)+\lambda y(x)=x^2+1+\lambda (x-2)(x-4)$$

The Lagrange dual function $g(\lambda)$ can be written as follows:

$$g(\lambda)=\min L(x,\lambda)=\min z(x)+\lambda \min y(x)=\min \{ x^2+1+\lambda (x-2)(x-4) \}$$

Note that the Lagrangian dual function $g$ is concave because it is a pointwise minimum of linear functions.

The dual problem can be written as
$$\max g(\lambda) \quad \text{s.t.} \quad \lambda \geq 0$$

Note, that in case of the primal problem the optimal value for $x=2$ satifies the constraint y(2) as an eqaulity, i.e. Slater's condition holds. Slater's condition being satisfied means that **strong duality** holds. Thus the optimal solution for the primal problem equals to the optimal solution of the dual problem.
